# 🤖  Čo je Machine Learning?
- Oblasť umelej inteligencie, ktorá sa zaoberá vývojom algoritmov a modelov, ktoré umožňujú počítačom učiť sa z údajov a zlepšovať svoje výkony bez explicitného programovania 
- Modely sa trénujú na historických dátach, aby dokázali predpovedať alebo klasifikovať nové, neznáme údaje

## Typy Strojového Učenia
**A. Učenie s učiteľom (Supervised Learning)**: Model sa trénuje na dátach, ktoré obsahujú vstupy a zodpovedajúce výstupy. Cieľom je naučiť model predpovedať výstupy na základe nových vstupov. Príkladmi sú regresia a klasifikácia.  
**B. Učenie bez učiteľa (Unsupervised Learning)**: Model sa trénuje na dátach bez explicitných výstupov. Cieľom je objaviť vzory alebo štruktúry v údajoch, ako je zhlukovanie alebo redukcia dimenzií.  
**C. Učenie so zosilnením (Reinforcement Learning)**: Model sa učí na základe skúseností a spätnej väzby z prostredia. Cieľom je optimalizovať sériu akcií, aby sa dosiahol maximálny zisk.  

# ⚒️ Top Machine Learning knižnice
1. **🤖 scikit-learn**: Knižnica pre Python, ktorá poskytuje **jednoduché** a **efektívne** **nástroje** na analýzu dát a strojové učenie. Obsahuje rôzne algoritmy pre klasifikáciu, regresiu a zhlukovanie.
   
2. **📊 TensorFlow**: Otvorená knižnica od **Google** pre numerické výpočty a strojové učenie. Je obzvlášť populárna pre hlboké učenie a podporuje rozsiahle nasadenie na rôznych platformách.

4. **🐍 PyTorch**: Otvorená knižnica od **Facebooku** pre hlboké učenie. Umožňuje dynamické výpočtové grafy a je známa svojou intuitívnosťou a jednoduchosťou pri vývoji komplexných modelov.

## Klasifikačný report pre klasifikátor
- Funkcia classification_reportvytvára textovú správu zobrazujúcu hlavné klasifikačné metriky

1. **Precision (presnosť)**: Miera správnych pozitívnych predpovedí k všetkým predpovediam pozitívnych tried. 
   
2. **Recall (citlivosť)**: Miera správnych pozitívnych predpovedí k všetkým skutočným pozitívnym hodnotám. 
   
3. **F1-score**: Harmonický priemer presnosti a citlivosti. 
   
4. **Support (podpora)**: Počet skutočných prípadov v testovacej sade pre každú triedu.


- **Accuracy (presnosť)**: Celková miera správnych predpovedí modelu (0.97) naznačuje, že model funguje veľmi dobre na celej testovacej sade.
- **Macro avg**: Priemerné hodnoty pre presnosť, citlivosť a F1-score vypočítané bez váženia tried, čo znamená, že každá trieda má rovnakú váhu.
- **Weighted avg**: Priemerné hodnoty, ktoré zohľadňujú počet prípadov v jednotlivých triedach, čo je dôležité, ak sú triedy nevyvážené.

# Hierarchický klastrový dendrogram

Dendrogram je diagram, ktorý zobrazuje hierarchickú štruktúru zhlukovania alebo klasifikácie údajov. Používa sa najmä v analýze zhlukovania na vizualizáciu spôsobu, akým sú objekty zoskupené do klastrov.

Hlavné vlastnosti dendrogramu:
Os x: Predstavuje jednotlivé objekty alebo vzorky.
Os y: Zobrazuje vzdialenosti alebo podobnosti medzi objektmi.
Zväzovanie: Zobrazuje, ako sú objekty spojené do klastrov, pričom čím vyššia je hodnota na os y, tým je zhlukovanie menej podobné.
Vytváranie zhlukov: Dendrogram umožňuje zvoliť úroveň, pri ktorej sa klastre delia, čo môže pomôcť pri identifikácii optimálneho počtu klastrov.

## Jednoduché imputovanie chýbajúcich hodnôt
- scikit-learn ponúka triedu SimpleImputer na náhradu chýbajúcich hodnôt. 
- Táto trieda umožňuje nahradiť chýbajúce hodnoty rôznymi spôsobmi, ako je priemer, medián alebo najčastejšia hodnota.
- V tomto príklade chýbajúce hodnoty v stĺpcoch Vek a Nákupy sú nahradené priemerom týchto stĺpcov.

## 🏘️ K-najbližších susedov imputácia
- Trieda KNNImputer je rozšírenou verziou imputácie, ktorá používa algoritmus k-najbližších susedov na vyplnenie chýbajúcich hodnôt. 
- Každá chýbajúca hodnota je nahradená priemerom hodnôt susedných bodov.
- V tomto príklade KNNImputer vypočíta chýbajúce hodnoty na základe priemeru hodnôt od 2 najbližších susedov.

## ⚡ Odstraňovanie outlierov pomocou robustnej škálovacej metódy
- RobustScaler škáluje dáta pomocou mediánu a kvartilov, čo je odolné voči outlierom (extrémnym hodnotám). 
- Nie je to priamo čistenie dát, ale pomáha minimalizovať vplyv outlierov.
- RobustScaler pomáha redukovať efekt extrémnych hodnôt bez ich odstránenia.


## 🤕 Odstraňovanie veľmi slabých alebo nekorelovaných čŕt:
- Trieda VarianceThreshold je jednoduchý spôsob, ako odstrániť črty (stĺpce), ktoré majú veľmi nízku variabilitu, a preto neprispievajú k modelovaniu.
- V tomto prípade je stĺpec Vek odstránený, pretože má nulovú variabilitu (všetky hodnoty sú rovnaké).

# 🤖 Normalizácia dát
- Normalizácia dát je proces, ktorý zabezpečuje, že hodnoty v rôznych stĺpcoch sú na porovnateľnej škále. 
- To je dôležité najmä pri analýze alebo trénovaní modelov strojového učenia.

## Min-Max Normalizácia
- Tento prístup škáluje hodnoty do rozmedzia [0, 1].
- Min-Max normalizácia je technika škálovania dát, ktorá transformuje hodnoty do zadaného rozsahu (zvyčajne medzi 0 a 1). 
- Jej cieľom je previesť rôzne hodnoty z rôznych rozsahov do jednotného rozsahu, čo môže byť dôležité napríklad pri strojovom učení, kde algoritmy fungujú lepšie s normalizovanými dátami.

## Z-score Normalizácia (Standardizácia)
- Tento prístup odčíta priemer a delí výsledok štandardnou odchýlkou, čím zabezpečuje, že výsledky majú priemer 0 a štandardnú odchýlku 1.
- Z-score (štandardizovaný skóre) je metóda používaná na vyjadrenie, o koľko štandardných odchýlok je hodnota od priemeru danej množiny dát. 
- Inými slovami, Z-score ukazuje, ako ďaleko (v pozitívnom alebo negatívnom smere) sa nachádza konkrétna hodnota v porovnaní s priemerom dát.

## Normalizácia pomocou logaritmu
- Logaritmická normalizácia môže byť užitočná pre dáta s exponenciálnym rozdelením.
- Logaritmická normalizácia je technika používaná na transformáciu dát, najmä v prípadoch, keď sú hodnoty rozdelené nerovnomerne, alebo existujú veľké rozdiely medzi jednotlivými hodnotami. 
- Tento prístup znižuje vplyv veľkých hodnôt a umožňuje lepšie pracovať s dátami, ktoré majú exponenciálny charakter alebo obsahujú extrémne odľahlé hodnoty (outliery).
