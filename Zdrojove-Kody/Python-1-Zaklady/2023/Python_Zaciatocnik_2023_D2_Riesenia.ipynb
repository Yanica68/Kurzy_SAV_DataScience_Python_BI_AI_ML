```python
# pexeso 
# pexeso game

import random

# Define pairs 
pairs = ['A', 'B', 'C', 'D']

# Shuffle the deck
deck = pairs * 2
random.shuffle(deck)

# The game state
state = ['_'] * 8

while '_' in state:
    print(' '.join(state))
    first_choice = int(input("Enter the first index (between 0 to 7): "))
    print(deck[first_choice])
    second_choice = int(input("Enter the second index (between 0 to 7): "))
    if deck[first_choice] == deck[second_choice]:
        state[first_choice] = deck[first_choice]
        state[second_choice] = deck[second_choice]
    else:
        print("Sorry, try again.")
        
print(' '.join(state))
print("Congratulations! You've matched all pairs.")
```

```python
zoznam_studenti = ["Maros", "Monika", "Peter", "Jozef", "Zuzana", "Eva"]

print("\nZoznam Reverzny Otoceny")
for student in reversed(zoznam_studenti):
    print(student)

print("\nCislovany Zoznam Reverzny Otoceny")
for i, student in enumerate(reversed(zoznam_studenti)):
    print(f"{i+1} {student}")

print("\nZoznam Zoradeny")
for student in sorted(zoznam_studenti, reverse=True):
    print(student)
```

```python
# fraktal  
# fraktal
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

def draw_mandelbrot(xmin,xmax,ymin,ymax,width,height,max_iter):
    r1 = np.linspace(xmin, xmax, width)
    r2 = np.linspace(ymin, ymax, height)
    return (r1,r2,np.array([[mandelbrot(complex(r, i),max_iter) for r in r1] for i in r2]))

image_width = 800
image_height = 800
xmin = -2.0
xmax = 1.0
ymin = -1.5
ymax = 1.5
max_iter = 256

dx, dy, mandelbrot_set = draw_mandelbrot(xmin,xmax,ymin,ymax,image_width,image_height,max_iter)

plt.imshow(mandelbrot_set, extent=(xmin, xmax, ymin, ymax))
plt.show()
```

```python
# vytvor eshop 
class Product:
    def __init__(self, id, name, price):
        self.id = id
        self.name = name
        self.price = price

class ShoppingCart:
    def __init__(self):
        self.items = []
        
    def add_item(self, product):
        self.items.append(product)
        
    def checkout(self):
        total = sum([item.price for item in self.items])
        self.items = []  # Clear cart after checkout
        return total

# Create some products
product1 = Product(1, 'Apple', 1.0)
product2 = Product(2, 'Orange', 1.5)

# Add products to shopping cart
cart = ShoppingCart()
cart.add_item(product1)
cart.add_item(product2)

# Checkout
total = cart.checkout()
print('Total: ', total)
```
